```python
# kfold_train.py
import os
from ultralytics import YOLO

# ====== ä¾ä½ çš„å°ˆæ¡ˆèª¿æ•´é€™è£¡ ======
ROOT_DIR   = "/NFS/celine/aicup/yoloNew"
KFOLDS     = 5
MODEL_PATH = "yolo12s.pt"     # ä½ çš„é è¨“ç·´æ¬Šé‡
EPOCHS     = 200
IMGSZ      = 512
BATCH      = 16
DEVICE     = 0                 # å–®å¡ 0ï¼›å¤šå¡å¯æ”¹æˆ [0,1] æˆ– '0,1'
LR0        = 0.003         # ä½ å¯¦é©—ä¸­ç”¨éçš„ lr0
OUTPUT_PROJ= os.path.join(ROOT_DIR, "runs", "kfold_train")
SINGLE_CLS = True
OPTIM      = "SGD"           # æˆ– 'SGD'ã€'Adam' ç­‰
SEED       = 42
# =================================

os.makedirs(OUTPUT_PROJ, exist_ok=True)

for k in range(KFOLDS):
    data_yaml = os.path.join(ROOT_DIR, f"kfold_{KFOLDS}", f"fold{k}", f"data_fold{k}.yaml")
    name = f"fold{k}"

    model = YOLO(MODEL_PATH)
    print(f"=== é–‹å§‹è¨“ç·´ fold {k} ===")
    model.train(
        data=data_yaml,
        epochs=EPOCHS,
        imgsz=IMGSZ,
        batch=BATCH,
        device=DEVICE,
        name=name,
        project=OUTPUT_PROJ,
        seed=SEED,
        single_cls=SINGLE_CLS,
        optimizer=OPTIM,
        lr0=LR0,
        patience=20,
        verbose=False,
        # å…¶ä»–ä½ æ—¢æœ‰çš„åƒæ•¸...
    )
    print(f"=== çµæŸ fold {k} ===")
```
