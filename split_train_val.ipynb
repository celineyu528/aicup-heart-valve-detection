把官方給的訓練資料(celine/aicup/yoloNew/images)分成train和val資料夾，labels也是。
跑完此程式，yoloNew資料夾會多跑出train和val資料夾，裡面各自有images資料夾和labels資料夾。

```python
import os
import random
import shutil

# 原始影像資料夾
SRC_DIR = '/NFS/celine/aicup/yoloNew/images'
LABEL_DIR = '/NFS/celine/aicup/yoloNew/labels'

# 目標資料夾
TRAIN_IMG_DIR = '/NFS/celine/aicup/yoloNew/train/images'
VAL_IMG_DIR = '/NFS/celine/aicup/yoloNew/val/images'
TRAIN_LABEL_DIR = '/NFS/celine/aicup/yoloNew/train/labels'
VAL_LABEL_DIR = '/NFS/celine/aicup/yoloNew/val/labels'

os.makedirs(TRAIN_IMG_DIR, exist_ok=True)
os.makedirs(VAL_IMG_DIR, exist_ok=True)
os.makedirs(TRAIN_LABEL_DIR, exist_ok=True)
os.makedirs(VAL_LABEL_DIR, exist_ok=True)

# 取得所有影像清單
images = [f for f in os.listdir(SRC_DIR) if f.endswith(('.jpg', '.png'))]
random.shuffle(images)

# 設定驗證集比例
val_ratio = 0.1
val_count = int(len(images) * val_ratio)

val_images = images[:val_count]
train_images = images[val_count:]

def move_files(file_list, dst_img_dir, dst_label_dir):
    for img_file in file_list: #逐一處理清單中的每個影像檔案
        base = os.path.splitext(img_file)[0]
        label_file = base + '.txt' #把影像檔名對應到標籤檔名
        shutil.copy(os.path.join(SRC_DIR, img_file), os.path.join(dst_img_dir, img_file)) 
        #os.path.join() 用來安全地組合檔案路徑
        if os.path.exists(os.path.join(LABEL_DIR, label_file)): 
            #檢查對應的標籤檔是否存在。（有些圖片可能沒有標籤，例如測試資料）
            shutil.copy(os.path.join(LABEL_DIR, label_file), os.path.join(dst_label_dir, label_file))
            #如果標籤存在，則一併複製到目標標籤資料夾

move_files(train_images, TRAIN_IMG_DIR, TRAIN_LABEL_DIR)
move_files(val_images, VAL_IMG_DIR, VAL_LABEL_DIR)

print(f"訓練集影像數量: {len(train_images)}")
print(f"驗證集影像數量: {len(val_images)}")
```

定義一個函式 move_files()

file_list：要搬的影像檔清單（例如：['0001.png', '0002.png', ...]）

dst_img_dir：目標影像資料夾（例如：train/images）

dst_label_dir：目標標籤資料夾（例如：train/labels）

os.path.splitext() 會把檔名分成「主檔名」和「副檔名」。
例如：

img_file = '0001.png'
os.path.splitext(img_file)
# 回傳 ('0001', '.png')


所以 base 就是 '0001'。

用 shutil.copy() 複製檔案。

第一個參數 → 來源檔案（要被複製的地方）

第二個參數 → 目的檔案（要被放到哪裡）

os.path.join(SRC_DIR, img_file)：來源路徑，例如 /NFS/.../images/0001.png

os.path.join(dst_img_dir, img_file)：目的路徑，例如 /NFS/.../train/images/0001.png

os.path.join(SRC_DIR, img_file)

意思是：「把資料夾 SRC_DIR 和 檔名 img_file 組合成完整路徑」。

假設：

SRC_DIR = '/NFS/celine/aicup/yoloNew/images'

dst_img_dir = '/NFS/celine/aicup/yoloNew/train/images'

img_file = '0001.png'


那：

| 指令                                    | 結果                                                |
| ------------------------------------- | ------------------------------------------------- |
| `os.path.join(SRC_DIR, img_file)`     | `/NFS/celine/aicup/yoloNew/images/0001.png`       |
| `os.path.join(dst_img_dir, img_file)` | `/NFS/celine/aicup/yoloNew/train/images/0001.png` |
